# 6. ベイズニューラルネットワークによるベイズ最適化

ベイズ最適化の関数モデリングは一般にガウス過程回帰モデルが用いられています。しかし、ニューラルネットワークのパラメータのベイズ推定がフレームワークとして確立され「ベイズニューラルネットワーク」とよばれるようになってから、ベイズニューラルネットワークをベイズ最適化の関数モデリングとして応用することも増えてきました。ここでは、その一例であるBOHAMIANNモデルを`emukit`パッケージを用いて計算してみましょう。

```python
import numpy as np
import matplotlib.pyplot as plt

from emukit.core import ParameterSpace, ContinuousParameter
from emukit.bayesian_optimization.acquisitions import ExpectedImprovement
from emukit.core.optimization import GradientAcquisitionOptimizer

import torch
import torch.nn as nn
from emukit.examples.models.bohamiann import Bohamiann
```

次の関数の最小値を求める問題をベイズ最適化で解いてみましょう。この関数はForrester関数といって、最適化のデモによく用いられる関数のひとつのです。

$$
\begin{align*}
f(x) &= (6x-2)^2\sin(12x-4), 0\leq x\leq1
\end{align*}
$$

```python
def forrester(x):
    return (6*x-2)**2 * np.sin(12*x - 4)
```

```python
x = np.linspace(start = 0, stop = 1, num = 100)
plt.plot(x, forrester(x))
plt.show()
```

```python
X_init = np.linspace(start = 0.05, stop = 0.95, num = 5)[:, np.newaxis]
Y_init = forrester(X_init)
```

## 6.2　変数の指定

まず最適化したい変数の名前、連続変数かカテゴリー変数か、値の範囲を指定します。

```python
x = ContinuousParameter(name = "x", min_value = 0, max_value = 1)
space = ParameterSpace(parameters = [x])
```

## 6.3　モデリング

今回は、4層のニューラルネットワーク（5->10->10->1）を用います。一般のニューラルネットワークと異なり、ベイズニューラルネットワークでは事後分散に対応する出力を作る必要があります。今回は`AppendLayer`クラスを実装し、`Sequential` apiでアーキテクチャに結合します。なお、初期値の指定を明示的に行うことで、事前分散を指定します。

```python
def get_network(input_dimensionality):
    class AppendLayer(nn.Module):     # 誤差の分散の実装
        def __init__(self, bias=True, *args, **kwargs):
            super().__init__(*args, **kwargs)
            if bias:
                self.bias = nn.Parameter(torch.Tensor(1, 1))
            else:
                self.register_parameter('bias', None)

        def forward(self, x):
            return torch.cat((x, self.bias * torch.ones_like(x)), dim = 1)

    def init_weights(module):    # 各パラメータの初期値の設定（特にAppendLayerの初期値は大事）
        if type(module) == AppendLayer:
            nn.init.constant_(module.bias, val = np.log(1))    # val引数に渡される値が事前分散（の対数）
        elif type(module) == nn.Linear:
            nn.init.kaiming_normal_(module.weight, mode = "fan_in", nonlinearity = "linear")
            nn.init.constant_(module.bias, val = 0.0)

    return nn.Sequential(    # ベイズニューラルネットワークの実装
        nn.Linear(input_dimensionality, 10), nn.ReLU(),
        nn.Linear(10, 10), nn.ReLU(),
        nn.Linear(10, 1),
        AppendLayer()
    ).apply(init_weights)
```

考えたアーキテクチャを`Bohamiann`クラスに渡し、事後平均と事後分散を求めておきます。

```python
emukit_model = Bohamiann(X_init, Y_init, get_architecture = get_network)
```

## 6.4　獲得関数の指定

今回はEIを用いましょう。`ExpectedImprovement`クラスに学習済みモデルを渡し、獲得関数の具体的な式を求めます。

```python
ei_acquisition = ExpectedImprovement(emukit_model)
```

## 6.5　実験候補の導出

後は、獲得関数の最大値を勾配降下法で求めます。

```python
optimizer = GradientAcquisitionOptimizer(space)
x_new, _ = optimizer.optimize(ei_acquisition)
x_new
```

## 6.6　確認

得られたモデルと獲得関数を出力しておきましょう。

```python
x_plot = np.linspace(start = 0, stop = 1, num = 100)[:, np.newaxis]
y_plot = forrester(x_plot)
mu_plot, var_plot = emukit_model.predict(x_plot)

plt.figure(figsize=(12, 8))
plt.plot(X_init, Y_init, "ro", markersize=10, label="Observations")
plt.plot(x_plot, y_plot, "k", label="Objective Function")
plt.plot(x_plot, mu_plot, "C0", label="Model")
plt.fill_between(x_plot[:, 0],
                 mu_plot[:, 0] + 1.96 * np.sqrt(var_plot)[:, 0],
                 mu_plot[:, 0] - 1.96 * np.sqrt(var_plot)[:, 0], alpha=0.6)
plt.legend(loc = 2)
plt.xlabel(r"$x$")
plt.ylabel(r"$f(x)$")
plt.xlim(0, 1)
plt.show()
```

```python
ei_plot = ei_acquisition.evaluate(x_plot)

plt.figure(figsize=(12, 8))
plt.plot(x_plot, (ei_plot - np.min(ei_plot)) / (np.max(ei_plot) - np.min(ei_plot)))
plt.axvline(x_new, color="red", label="x_next", linestyle="--")
plt.legend(loc = 1)
plt.xlabel(r"$x$")
plt.ylabel(r"$f(x)$")
plt.xlim(0, 1)
plt.show()
```
